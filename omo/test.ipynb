```python
import pandas as pd
import json
```

```python
df = pd.read_csv('coefs2.csv')
```

```python
variables = df['decision_variables_xijk'].tolist()
```

```python
df.head()
```

```python
objectives = [df["obj_ponto_inicial"].tolist(), df["obj_tempo_viagem"].tolist()]
```

```python
restrictions_dict = df.iloc[:, 3:]
```

```python
restrictions = []
for column in restrictions_dict:
    tmp_var = restrictions_dict[column].tolist()
    restrictions.append(tmp_var)
```

```python
constraints_sign_list = [
    "=1",
    "=1",
    "=1",
    "=1",
    "=1",
    "=1",
    "=1",
    "<=2",
    "<=4",
    "=0",
    "=0",
    "=0",
    "=0",
    "=0",
    "=0"
]
```

```python
newJson = {
    "variables": variables,
    "objectives": objectives,
    "constraints": restrictions,
    "constraint_types": constraints_sign_list
}
```

```python
with open('problemCoefs.json', 'w') as outfile:
    json.dump(newJson, outfile)
```
